# Intialization & Data Cleaning

```python
# Import pandas
import pandas as pd
```

```python
# Create dataframe
df = pd.read_csv("fatalities_isr_pse_conflict_2000_to_2023.csv")
```

```python
# Create an overview of the data
df.head(10)
```

```python
# Drop empty values and remove duplicates
df = df.dropna()
df = df.drop_duplicates()
```

```python
# Remove columns unneccessary for analysis
df = df.drop(columns=['name', 'notes', 'place_of_residence', 'place_of_residence_district'])
```

```python
# View the current dataframe
df.head(10)
```

```python
# Round age to a whole integer with no decimal
df['age'] = df['age'].astype(int)
```

```python
# 'date_of_event' and 'date_of_death' are redundant and thus one can be removed
df= df.drop(columns=['date_of_event'])
```

```python
# Remove information on events not located in Gaza
df = df[df['event_location_district'].str.contains('Gaza')]
```

```python
# Reset index
df = df.reset_index(drop=True)
```

# Exploratory Data Analysis

In the subsequent section, I create an Exploratory Data Analysis (EDA) of the Israel-Palestine conflict in Gaza. EDA is a pivotal step in data analysis, providing a visual and quantitative examination of the dataset's main patterns, relationships, anomalies, and structures.

Beginning with an age distribution, we can see that most victims in this conflict are young, typically in their twenties. Following this, a time-series analysis portrays the frequency of fatal events over an extended period, shedding light on the temporal patterns, peaks, and troughs of violent incidents. Lastly, I assess the distribution of participation in the conflict, categorizing participants based on their involvement levels with a pie chart. This visualization offers a comprehensive understanding of the distinct roles individuals have played within the context of the conflict.

As I navigate through the EDA, my objective is not only to provide a clearer picture of the current situation but also to identify trends and nuances that could inform future visualizations in Tableau. Tableau is a great tool for creating visually appealing charts and diagrams that are easily interpreted by stakeholders and individuals not from a data background.

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the style for seaborn
sns.set_style("whitegrid")

# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by date and count the number of events on each day
time_series_data = df.groupby('date_of_death').size()

# Plot the time series data
plt.figure(figsize=(12, 6))
time_series_data.plot()
plt.title('Time-Series Analysis of Fatal Events Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.tight_layout()
plt.show()
```

```python
# Create a pie chart of active participants in the conflict and bystanders

# Group by 'took_part_in_the_hostilities' and count the occurrences
participation_counts = df['took_part_in_the_hostilities'].value_counts()

# Plotting
fig, ax = plt.subplots(figsize=(12, 7))

# Using a custom color palette
colors = ['lightblue', 'orange', 'lightgreen', 'pink', 'gray']

participation_counts.plot(kind='pie', colors=colors, autopct='%1.1f%%', startangle=90, legend=False, ax=ax)
ax.set_ylabel('')  # This removes the default label to make the plot cleaner
plt.title('Distribution of Participation in the Conflict')

# Position the legend
ax.legend(title='Participation Type', bbox_to_anchor=(1, 0.5), loc='center left')

plt.show()
```
